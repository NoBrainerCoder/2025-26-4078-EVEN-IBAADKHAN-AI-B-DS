```python

import pandas as pd

data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [20, 40, 35, 45],
    "Age": [18, 20, 19, 18]
}

df = pd.DataFrame(data)
print("Original Data")
print(df)

df["Bonus"] = 5
print("After adding Bonus")
print(df)

df["Marks"] = df["Marks"] + 2
print("After increasing Marks")
print(df)

df["Total"] = df["Marks"] + df["Bonus"]
print("After adding Total")
print(df)
```

```python
# Selection sort
n = int(input("enter number of elements:"))

arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

print("\n Original array:", arr)

# selection sort logic
for i in range(n-1):
    min_index = i
    for j in range(i+1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    # swap found min element with first element
    arr[i], arr[min_index] = arr[min_index], arr[i]

print("sorted array:", arr)
```

```python
# Bubble sort
n = int(input("enter num of elements:"))

arr = []

for i in range(n):
    arr.append(input(input(f"enter element {i+1}:")))

print("\n Original array:", arr)

# Bubble sort logic
for i in range(n-1):
    for j in range(n-1-i):
        if arr[j] > arr[j+1]:
            # Swap
            arr[j], arr[j+1] = arr[j+1], arr[j]

print("Sorted array:", arr)
```

```python
# linear search

n = int(input("enter no. of elements: "))
arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i}: ")))

key = int(input("enter element to search: "))

found = False

for i in range(n):
    if arr[i] == key:
        print(f"element {key} found at position {i+1}")
        found = True
        break

if not found:
    print("element not found")
```
